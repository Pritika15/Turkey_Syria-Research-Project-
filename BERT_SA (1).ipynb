```python
import pandas as pd
```

```python
df=pd.read_csv('textBlob6lakh.csv')
```

```python
df
```

```python
#Label as 0-Negative, 1-Neutral, 2-Positive
label=[]
for senti in df['sentiment']:
  if(senti=='positive'):
    label.append(2)
  elif(senti=='negative'):
    label.append(0)
  else:
    label.append(1)    
label
```

```python
df['label']=label
```
