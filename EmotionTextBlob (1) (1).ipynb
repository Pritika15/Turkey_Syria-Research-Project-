```python
df=pd.read_csv('6lakhdata.csv')
df.head()
```

```python
Polarity=[]
subjectivity=[]
for textdata in df["text"]:
  sent=TextBlob(str(textdata))
  Polarity.append(sent.sentiment[0])
  subjectivity.append(sent.sentiment[1])
print("Polarity:")  
Polarity

```

```python
print("subjectivity:")  
subjectivity   
```

```python
df["label"]=Polarity
df['subjectivity']=subjectivity
```

```python
df.head(50)
```

```python
emotions=[]
for polarityScore in df['label']:
  if polarityScore > 0.3:
    emotions.append('relief')
  elif polarityScore > 0:
    emotions.append('happy')
  elif polarityScore < -0.3:
    emotions.append('disgust')
  elif polarityScore < 0:
    emotions.append('sad')
  else:
    emotions.append('neutral')
```

```python
df['emotion']=emotions
```

```python
df.head(50)
```

```python
import matplotlib.pyplot as plt
df['emotion'].value_counts().plot.pie(figsize=(6,6),title="Distribution of emotions",labels=['','','','',''],autopct='%1.1f%%')
labels=["relief","happy","disgust","sad","neutral"]
plt.legend(labels,loc=3)
```
