```python
import pandas as pd
```

```python
df=pd.read_csv('6lakhdata.csv')
df.head()
```

```python
import nltk
from nltk.sentiment import SentimentIntensityAnalyzer
```

```python
nltk.download('vader_lexicon')
```

```python
sid = SentimentIntensityAnalyzer()
```

```python
polarity=[]
for textVal in df['text']:
  scores = sid.polarity_scores(textVal)
  polarity.append(scores['compound'])

polarity  
```

```python
df['polarity']=polarity
```

```python
df.head()
```

```python
emotions=[]
for polarityScore in df['polarity']:
  if polarityScore >= 0.5:
    emotions.append('relief')
  elif polarityScore > 0:
    emotions.append('happy')
  elif polarityScore <= -0.5:
    emotions.append('disgust')
  elif polarityScore < 0:
    emotions.append('sad')
  else:
    emotions.append('neutral')
```

```python
df['emotion']=emotions
```

```python
df.head(50)
```

```python
import matplotlib.pyplot as plt
df['emotion'].value_counts().plot.pie(figsize=(6,6),title="Distribution of emotions",labels=['','','','',''],autopct='%1.1f%%')
labels=["relief","happy","disgust","sad","neutral"]
plt.legend(labels,loc=3)
```
